```python
import pandas as pd
```

```python
climer_data = pd.read_csv("/private/home/arjunrs1/CliMer/data/egoexo4d/egoexo4d_all_views_keysteps_train.csv")
ks_data = pd.read_csv("/private/home/arjunrs1/exo_narration_grounding/data_processing/time_interval_annotation_files/keystep_annotations/train.csv")
```

```python
# Function to extract take_uid from video_id
def extract_take_uid(video_id):
    rsplit_idx = 2 if "aria" in video_id else 1
    return video_id.rsplit('_', rsplit_idx)[0]
# Apply the function to extract take_uid
climer_data['take_uid'] = climer_data['video_id'].apply(extract_take_uid)
# Ensure data types match
climer_data['narration_frame'] = climer_data['narration_frame'].astype(int)
ks_data['narration_frame'] = ks_data['narration_frame'].astype(int)
# Strip whitespace and ensure case consistency
climer_data['take_uid'] = climer_data['take_uid'].str.strip().str.lower()
ks_data['take_uid'] = ks_data['take_uid'].str.strip().str.lower()
# Merge with indicator
merged_data = pd.merge(climer_data, ks_data, on=['narration_frame', 'take_uid'], how='inner', suffixes=('', '_y'), indicator=True)
# Filter to keep only rows that were successfully merged
climer_data_filtered = merged_data[merged_data['_merge'] == 'both'].drop(columns=['_merge'])
# Select the relevant columns
result = climer_data_filtered[['clip_id', 'unique_narration_id']]
print(result)
```
